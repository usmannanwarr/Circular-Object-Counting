```python
import os
import random
from PIL import Image, ImageDraw

# Create a directory to store the generated images
output_directory = "generated_images"
os.makedirs(output_directory, exist_ok=True)

# Number of images to generate
num_images = 50

# Image size
image_size = (256, 256)

# Max Number of circles per image
min_circles = 15
max_circles = 20

# Range of circle radii
min_radius = 7
max_radius = 10

def is_circle_overlapping(existing_circles, center, radius):
    for (existing_center, existing_radius) in existing_circles:
        distance = ((center[0] - existing_center[0])**2 + (center[1] - existing_center[1])**2)**0.5
        if distance < radius + existing_radius:
            return True
    return False

for i in range(num_images):
    # Create a new image with a white background
    img = Image.new("RGB", image_size, "white")
    draw = ImageDraw.Draw(img)

    # Draw random circles
    drawn_circles = []
    num_circles = random.randint(min_circles, max_circles)
    for _ in range(num_circles):
        # Try to find a non-overlapping circle
        for _ in range(100):
            radius = random.uniform(min_radius, max_radius)
            center = round(random.uniform(radius, image_size[0]-radius), 0), round(random.uniform(radius, image_size[1]-radius), 0)

            if not is_circle_overlapping(drawn_circles, center, radius):
                drawn_circles.append((center, radius))
                draw.ellipse([(center[0] - radius, center[1] - radius),
                              (center[0] + radius, center[1] + radius)], fill=(0, 0, 0), outline=(128,128,128))
                break

    # Save the image
    image_filename = os.path.join(output_directory, f"image_{i + 1}.png")
    img.save(image_filename)
```
